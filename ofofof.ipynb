```python
# voit objets image avec matchc?? + coord
for c in contours:
    matchc = cv.matchShapes(contours_A[0], c, 1, 0.0)
    # print(f"Match value: {matchc}")

    # Object recognition based on matchShapes value
    if 0.54 < matchc < 0.55:
        object_type = "A"
        color = (0, 255, 0)  # Green
    elif 0.5 < matchc < 0.5058:
        object_type = "B"
        color = (255, 0, 0)  # Blue
    elif 0.585 < matchc < 0.59:
        object_type = "C"
        color = (0, 0, 255)  # Red
    elif cv.contourArea(c) > 560 and cv.contourArea(c) < 1200:
        object_type = "fenetre"
        color = (0, 102, 255)  # orrange
    elif 0.463 < matchc < 0.47:
        object_type ="s"
        color = (203,192,255)
    elif cv.contourArea(c) > 9000 and cv.contourArea(c) < 10000:
        object_type = "pat"
        print("p")
        color = (145,145,0)
    elif 0.77 < matchc < 0.8:
        object_type = "porte"
        color = (102, 0, 102)
    elif cv.contourArea(c) > 1300 and cv.contourArea(c) < 3000:
        object_type = "lavabo"
        # A et porte  lavabo
        color =  (0, 255, 255)

# 255,192,203
    else:
        object_type = "Unknown"
        color = (128, 128, 128)  # Gray



    # If object type is recognized, process it further
    if object_type != "Unknown":
        print(f"Detected object: {object_type}")

        # Calculate the centroid of the object
        M = cv.moments(c)
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])
        else:
            cX, cY = 0, 0  # Handle cases where division by zero could occur

        # Print the coordinates
        print(f"Object {object_type} detected at coordinates ({cX}, {cY})")

        # Draw the contour
        cv.drawContours(target, [c], -1, color, 2)

        # Label the object with its type and coordinates
        label = f"{object_type} ({cX}, {cY})"
        cv.putText(target, label, (cX - 10, cY - 10), cv.FONT_HERSHEY_SIMPLEX,
                   0.6, color, 2)

# Display the final image with marked objects and coordinates
cv2_imshow(target)
```

```python

rooms = []

# Room detection and bounding box calculation
room_counter = 1  # To give unique names to rooms
for c in contours:
    if 3000 < cv.contourArea(c) < 4000:  # Example range for room size
        # Calculate room bounding box
        x, y, w, h = cv.boundingRect(c)
        room_coordinates = {"name": f"Room {room_counter}", "x": x, "y": y, "w": w, "h": h}
        rooms.append(room_coordinates)
        room_counter += 1

        # Draw the room contour (no text)
        cv.drawContours(target, [c], -1, (0, 255, 0), 2)  # Green contour for rooms

# Object detection and association with rooms
for c in contours:
    matchc = cv.matchShapes(contours_A[0], c, 1, 0.0)

    # Object recognition
    if 0.54 < matchc < 0.55:
        object_type = "A"
        color = (0, 255, 0)  # Green
    elif 0.5 < matchc < 0.5058:
        object_type = "B"
        color = (255, 0, 0)  # Blue
    elif 0.585 < matchc < 0.59:
        object_type = "C"
        color = (0, 0, 255)  # Red
    elif  0.463 < matchc < 0.47:
         object_type = "S"
         color = (203,192,255)
    else:
        object_type = "Unknown"
        continue

    # Calculate the object's centroid
    M = cv.moments(c)
    if M["m00"] != 0:
        cX = int(M["m10"] / M["m00"])
        cY = int(M["m01"] / M["m00"])
    else:
        cX, cY = 0, 0

    # Check which room the object is in
    object_room = None
    for room in rooms:
        if (room["x"] <= cX <= room["x"] + room["w"] and
                room["y"] <= cY <= room["y"] + room["h"]):
            object_room = room["name"]
            break

    # Print object and its associated room
    if object_room:
        print(f"Object {object_type} detected in {object_room}")
    else:
        print(f"Object {object_type} detected outside any room")

    # Draw the object's contour (no text)
    cv.drawContours(target, [c], -1, color, 2)

# Display the final image
cv2_imshow(target)
```
